### Numpy code

```python
import numpy as np

# Example: Large matrices (adjust size as needed)
n = 7000  # For very large matrices, ensure you have enough RAM
A = np.random.rand(n, n).astype(np.float32)
B = np.random.rand(n, n).astype(np.float32)

C = np.dot(A, B)  # warm-up and Matrix multiplication

%timeit -r 2 -o np.dot(A, B)

print(f"Result shape: {C.shape}")
print(f"Result type: {C.dtype}")
```

```python
import torch
device = 'cuda'
A_gpu = torch.rand(n, n, device=device, dtype=torch.float32)
B_gpu = torch.rand(n, n, device=device, dtype=torch.float32)

def run_gpu():
    res = torch.matmul(A_gpu, B_gpu)
    torch.cuda.synchronize()
    return res

#Calentamiento
C = run_gpu()


print("Usando Pytorch")
time_gpu = %timeit -r 2 -o run_gpu()
print(f"Result shape: {C.shape}")
print(f"Result type: {C.dtype}")
```

Usando Pytorch

50.1 ms ± 122 μs per loop (mean ± std. dev. of 2 runs, 10 loops each)

Result shape: torch.Size([7000, 7000])

Result type: torch.float32


```python
import torch
N = 10**8 
device = 'cuda'

#Generar datos (0 a 1)
x = torch.rand(N, device=device)
y = torch.rand(N, device=device)

def pi_pytorch():
    aciertos = torch.sum(x**2 + y**2 < 1.0)
    return 4 * aciertos.item() / N

print("Tiempo de cálculo de pi usando Pytorch")
time_gpu = %timeit -r 2 -o pi_pytorch()

pi = pi_pytorch()
print("Resultado Pi:", pi)
```
